# **Text Cleaning**

```python
import nltk
nltk.download('punkt')
```

```python
nltk.download('wordnet')
```

```python
nltk.download('stopwords')
```

```python
import pandas as pd
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from bs4 import BeautifulSoup
import unicodedata
from nltk.stem import WordNetLemmatizer

def clean_text(sentence):
    if not isinstance(sentence, str):
        return ""
    text = sentence.lower()

    text = BeautifulSoup(text, 'html.parser').get_text()

    text = ' '.join([word for word in text.split() if not word.startswith('http')])

    text = ''.join([char for char in text if char not in string.punctuation + '’‘'])

    text = ''.join([i for i in text if not i.isdigit()])

    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(text)
    text = ' '.join([word for word in word_tokens if word.lower() not in stop_words])

    text = unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('utf-8')

    lemmatizer = WordNetLemmatizer()
    tokens = word_tokenize(text)
    text = ' '.join([lemmatizer.lemmatize(word) for word in tokens])

    return text
```

```python
data_bat['review_text']=data_bat['review text'].apply(clean_text)
```

# **Setting up the Pipeline and Trainning**

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.metrics import accuracy_score, make_scorer
import time
import sys
import pickle
```

```python
# Define a pipeline with a CountVectorizer and a classifier
pipeline = Pipeline([
    ('vect', CountVectorizer()),
    ('clf', None)  # Placeholder for the classifier
])

# Define parameters for grid search for each classifier
parameters = [
    {
        'clf': [LogisticRegression()],
        'clf__C': [0.1, 1, 10],
        'clf__solver': ['liblinear', 'lbfgs']
    },
    {
        'clf': [MultinomialNB()],
        'clf__alpha': [0.1, 0.5, 1.0]
    },
    {
        'clf': [SVC()],
        'clf__C': [1, 10, 100],
        'clf__kernel': ['linear', 'rbf']
    },
    {
    'clf': [RandomForestClassifier()],
    'clf__n_estimators': [50, 100, 200],
    'clf__max_depth': [None, 10, 20],
    'clf__min_samples_split': [2, 5, 10]
    },
    {
        'clf': [KNeighborsClassifier()],
        'clf__n_neighbors': [3, 5, 7],
        'clf__weights': ['uniform', 'distance']
    },
    {
        'clf': [DecisionTreeClassifier()],
        'clf__max_depth': [None, 10, 20],
        'clf__min_samples_split': [2, 5, 10]
    },
    {
        'clf': [AdaBoostClassifier()],
        'clf__n_estimators': [50, 100, 200],
        'clf__learning_rate': [0.1, 0.5, 1.0]
    },
    {
        'clf': [GradientBoostingClassifier()],
        'clf__n_estimators': [50, 100, 200],
        'clf__learning_rate': [0.05, 0.1, 0.2],
        'clf__max_depth': [3, 4, 5]
    },
    {
        'clf': [BaggingClassifier()],
        'clf__n_estimators': [10, 50, 100],
        'clf__max_samples': [0.5, 1.0],
        'clf__max_features': [0.5, 1.0]
    },
    {
        'clf': [ExtraTreesClassifier()],
        'clf__n_estimators': [50, 100, 200],
        'clf__max_depth': [None, 10, 20]
    }
]

# Define a scoring function
scoring = make_scorer(accuracy_score)

# Perform grid search for each classifier
for params in parameters:
    grid_search = GridSearchCV(pipeline, params, cv=5, verbose=1, n_jobs=-1, scoring=scoring)
    mlflow.sklearn.autolog(max_tuning_runs=None)
    with mlflow.start_run() as run:
        %time grid_search.fit(X_train, y_train)

        model_name = grid_search.best_estimator_.named_steps['clf'].__class__.__name__
        train_score = grid_search.score(X_train, y_train)
        test_score = grid_search.score(X_test, y_test)
        fit_time = grid_search.cv_results_['mean_fit_time'].mean()
        predict_time = grid_search.cv_results_['mean_score_time'].mean()
        model_size = sys.getsizeof(pickle.dumps(grid_search.best_estimator_))

        print("********** {} **********".format(model_name.lower()))
        print("Train Score: ", train_score)
        print("Test Score: ", test_score)
        print("Fit Time: ", fit_time)
        print("Predict Time: ", predict_time)
        print("Model Size: ", model_size)
```
